# Searching NaN data

```python
print("NaN data :",Ht_De_Ptn.isnull().values.sum())

### dropping nan values
Ht_De_Ptn.dropna(inplace=True)
print("shape after removing nan data :",Ht_De_Ptn.shape)
```

# Data Visualization

```python
import matplotlib.pyplot as Ht_De_Matplt
import seaborn as Ht_De_sabn
import warnings as Ht_De_wngs
Ht_De_wngs.filterwarnings('ignore')
```

```python
Ht_De_sabn.countplot(x ='target', data = Ht_De_Ptn, palette="viridis")
Ht_De_Matplt.title('Heart Disease Prediction')
Ht_De_Matplt.show()
```

```python
Ht_De_sabn.lineplot(x="trestbps", y="target", data=Ht_De_Ptn)
Ht_De_Matplt.title('columns "chol" vs "target"')
Ht_De_Matplt.show()
```

```python
from sklearn import preprocessing as Ht_De_Ptn_Pre
Ht_De_Ptn_LE = Ht_De_Ptn_Pre.LabelEncoder()
Ht_De_Ptn_Pre = Ht_De_Ptn.apply(Ht_De_Ptn_LE.fit_transform)
Ht_De_Ptn_Pre
```
